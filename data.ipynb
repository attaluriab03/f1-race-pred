```python
schedule2014 = fastf1.get_event_schedule(2014, include_testing=False)
for index, row in schedule2014.iterrows():
    event = row['EventName']
    trackLoc = row['Location']
    raceResults = fastf1.get_session(2014, event, 'R')
    raceResults.load(telemetry=False, weather=False, messages=False)
    df_race = raceResults.results
    df_race = df_race.drop(columns=['BroadcastName', 'TeamColor', 'HeadshotUrl', 'CountryCode'])
    df_race['PositionChange'] = df_race['GridPosition'] - df_race['Position']
    df_race['Year'] = 2014 
    df_race['Event'] = event
    df_race['TrackLocation'] = trackLoc
    df_race.to_csv(f'raceData/{trackLoc}-{event}-2014.csv')
```

```python
# confirm that all the csv's for each year have non-null time and position change columns
def verifyData():
    count = 0
    for fileName in os.listdir('raceData/'):
        df = pd.read_csv(f'raceData/{fileName}')
        timeNull = df[df['Time'].isna()].shape[0]
        posNull = df[df['PositionChange'].isna()].shape[0]
        if (timeNull == df.shape[0] or posNull == df.shape[0]):
            print(fileName)
            count += 1

    if count == 0:
        return True
    print(count)
    return False
verifyData()
```

<h2> All code below is test code </h2>

```python
years = list(range(2014, 2025))
raceData = defaultdict(list)
for year in years:
    # time.sleep(5)
    currYearSchedule = None
    if year < 2018:
        currYearSchedule = fastf1.get_event_schedule(year, include_testing=False, backend='ergast')
    else:
        currYearSchedule = fastf1.get_event_schedule(year, include_testing=False, backend='fastf1')
    for event in currYearSchedule['EventName']:
        raceResults = None
        if year < 2018:
            raceResults = fastf1.get_session(year, event, 'R', backend='ergast')
        else:
            raceResults = fastf1.get_session(year, event, 'R', backend='fastf1')
        raceResults.load(laps=False, telemetry=False, weather=False, messages=False)
        # time.sleep(1)
        df_race = raceResults.results
        df_race = df_race.drop(columns=['BroadcastName', 'TeamColor', 'HeadshotUrl', 'CountryCode'])
        df_race['PositionChange'] = df_race['GridPosition'] - df_race['Position']
        df_race['Year'] = year
        df_race['Event'] = event
        raceData[event].append(df_race)
```
